```python
import pandas as pd
from sklearn.preprocessing import normalize
import json
import msacorrnet as mcn
from copy import deepcopy
from sklearn.manifold import MDS
import matplotlib.cm as cm
from scipy.spatial import distance
import random
import matplotlib as mpl
from itertools import groupby
import topcorr
import fnmatch
import os
import glob
from tqdm import tqdm
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from collections import defaultdict
import itertools
from matplotlib.patches import Circle
from matplotlib.cm import get_cmap
```

```python
time = int(7200)

file = "data/hts-fault-2hr/cpu_usage.json"

f = open(file)

data = json.load(f)
f.close()

f_dict = {}
data = data['data']['result']
for ent in range(len(data)):
    if data[ent]['metric']['id'] != "/":
        f_dict[data[ent]['metric']['pod']] = np.zeros((len(data[ent]['values']), 2))
        f_dict[data[ent]['metric']['pod']][:, 0] = [float(x[0]) for x in data[ent]['values']]
        f_dict[data[ent]['metric']['pod']][:, 1] = [float(x[1]) for x in data[ent]['values']]

for service in f_dict:
    if data[0]["metric"]["__name__"] == "container_memory_usage_bytes":
        pass
    else:
        temp = [0]
        for i in range(1, np.shape(f_dict[service])[0]):
            temp.append(f_dict[service][i, 1] - f_dict[service][i - 1, 1])
        f_dict[service][:, 1] = deepcopy(temp)
        f_dict[service][:, 1] = mcn.sub_sample_zeros(f_dict[service][:, 1])

max_len = 0
for i in f_dict:
    if len(f_dict[i]) > max_len:
        max_len = len(f_dict[i])

aligned_data = mcn.align_data(f_dict, 5, time)
service_list = []
for s in aligned_data:
    service_list.append(s)
s_list = []
for s in aligned_data:
    s_list.append(s)
X = np.zeros((int((time / 5) ), len(s_list)))
ind = 0
for s in service_list:
    X[:, ind] = aligned_data[s][:, 1]
    ind += 1
```

```python
# remove misc data
misc_ms = ["nginx*","jaeger*"]
for ms in misc_ms:
    matched_indices = [index for index, string in enumerate(service_list) if fnmatch.fnmatchcase(string, ms)]
    if len(matched_indices) > 0:
        s_list.pop(matched_indices[0])
        service_list.pop(matched_indices[0])
        X = np.delete(X,matched_indices[0],axis=1)
```

```python
dir = os.getcwd()+'/data/hts-fault-2hr'
os.chdir(dir)
files = glob.glob('traces_*.json')
spans_list = []
for file in tqdm(files):
    with open(dir+"\\"+file, 'r') as file:
        data = json.load(file)
        
    spans = data['data']
    
    # Iterate through each trace and span to extract start time and duration
    for trace in data['data']:
        temp = {'startTime': 0, 'duration': 0}
        for span in trace['spans']:
            if span['duration'] > temp['duration']:
                temp['startTime'] = span['startTime']
                temp['duration'] = span['duration']
                
        spans_list.append(temp)

# Convert spans list to calls per second
df = pd.DataFrame(spans_list)
df['startTime'] = pd.to_datetime(df['startTime'], unit='us')
df['duration'] = df['duration'] / 1e3
df = df.set_index('startTime')
df = df[~df.index.duplicated(keep='first')] # remove overlapping traces
average_response_time_per_second = df['duration'].resample('5s').mean()
calls_per_second = df.resample('5s').size()
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1) 
average_response_time_per_second.plot()
plt.title('Average Response Time Per Second')
plt.xlabel('Time')
plt.ylabel('Average Response Time (ms)')
plt.subplot(3, 1, 2)
calls_per_second.plot()
plt.title('Number of Calls Per Second')
plt.xlabel('Time')
plt.ylabel('Number of Calls')
plt.subplot(3,1,3)
plt.plot(X)
plt.xlim(0,1440)
plt.title('CPU Usage')
plt.xlabel('Sample Number')
plt.ylabel('Normalized CPU Usage')
plt.tight_layout()
plt.show()
```

```python
win = 360
step = 1
Xc = deepcopy(X)
Xdiff = np.diff(Xc[:,:],axis=0)
Xdiff = normalize(Xdiff,axis=0)
r_dict = mcn.rolling_window(Xdiff, win, "tapered", step, 0.3, False, "pearsons")
```

```python
r_pos = {}
for corr in r_dict:
    r_pos[corr]=deepcopy(r_dict[corr])
    r_pos[corr] = r_pos[corr] + 1
    
spectra = mcn.all_spectra(r_pos, True)
all_es = spectra.T
all_es = all_es[:,:-1]
```

```python
scaling="LMDS"
if scaling == "LMDS":
    landmarks = random.sample(range(0, all_es.shape[0], 1), int(len(all_es) / 4))
    dist = distance.cdist(all_es[landmarks, :], all_es, 'euclidean')
    coords = mcn.LMDS(dist, landmarks, 2)

elif scaling == "MDS":
    dist = distance.cdist(all_es, all_es, 'euclidean')
    mds = MDS(n_components=2, dissimilarity="precomputed", random_state=0)
    coords = mds.fit_transform(dist)
```

```python
# Finalized plot
fig, ax = plt.subplot_mosaic('AAA;BBB;CCC', gridspec_kw=dict(
    height_ratios=[1, 0.1, 3]), dpi=150)

n = len(r_pos)
cmap_type = "viridis"
type_color = "ocean"
cmap = cm.get_cmap(type_color, n)
cmap = cmap.reversed()
t = np.arange(0, n)
plt.tight_layout()
fin = int(np.floor(len(X)/step)*step)
ax['A'].set_ylabel('CPU Usage')
ax['A'].set_title('CPU Timeseries of All Microservices')
ax['A'].set_xlabel("Time (s)")
ax['A'].plot(range(win*5,len(Xc)*5,5),normalize(Xc[win:,:]*100,axis=1), alpha=0.5, linewidth=1)
ax['A'].set_xlim(xmin=win*5)
ax['A'].set_xlim(xmax=len(X)*5)
cmap = cm.get_cmap(type_color, np.shape(X)[1])
colors = [cmap(i) for i in range(np.shape(X)[1])]
for i, j in enumerate(ax['A'].lines):
    j.set_color(colors[i])
    
ax['A'].axvline((720*5), color='k',linestyle='--')
cbar = plt.colorbar(mpl.cm.ScalarMappable(norm=mpl.colors.Normalize(int(win*5), time),
                                          cmap=cmap_type),
                    ax=ax['B'],
                    orientation='horizontal',
                    fraction=3.5,
                    pad=0,
                    )
ax['B'].set_visible(False)
cbar.ax.tick_params(size=0)
cbar.set_ticks([])
ax['C'].scatter(coords[:, 0], coords[:, 1], c=t, cmap=cmap_type, alpha=1)

n_pts = 25
if n_pts > 0:
    prev_x = np.mean(coords[0:n_pts, 0])
    prev_y = np.mean(coords[0:n_pts, 1])
    for i in range(10, len(coords), n_pts):
        curr_x = np.mean(coords[i:i+n_pts, 0])
        curr_y = np.mean(coords[i:i+n_pts, 1])
        ax['C'].plot([prev_x, curr_x], [prev_y, curr_y], 'k', alpha=0.5, linewidth=0.1)
        ax['C'].plot(prev_x, prev_y, 'k.', markersize=2, alpha=0.5)
        ax['C'].annotate("",
                              xytext=(prev_x, prev_y),
                              xy=(curr_x, curr_y),
                              arrowprops=dict(arrowstyle="->", color='k', alpha=0.5, linewidth=1))
        prev_x = curr_x
        prev_y = curr_y


plt.show()
```

```python
clustering = AgglomerativeClustering(n_clusters=2, linkage="average").fit(coords)
plt.scatter(coords[:, 0],coords[:, 1], c=clustering.labels_, cmap='rainbow', marker='o')
plt.title('Agglomerative Clustering')
plt.show()
```

```python
labels=clustering.labels_
# Find the order of first instance cluster appearance
order_of_clusters = [key for key, _ in groupby(labels)]

unique_clusters = []
seen = set()
for label in labels:
    if label not in seen:
        unique_clusters.append(label)
        seen.add(label)

network_centroids = {}
for i in range(0,2):
    avg_network = np.zeros((np.shape(r_dict[0])))
    for n in np.where(labels == unique_clusters[i])[0]:
        avg_network = avg_network+r_dict[n]
    avg_network = avg_network/len(np.where(labels == i))
    network_centroids[i] = deepcopy(avg_network)
```

```python
def extract_service_type(node_label):
    return '-'.join(node_label.split('-')[:2])


def initial_positions(service_clusters):
    pos = {}
    cluster_offset = 3  # Distance offset between clusters
    for cluster_index, (service_type, nodes) in enumerate(service_clusters.items()):
        angle_offset = cluster_index * (2 * np.pi / len(service_clusters))
        cluster_center = np.array([np.cos(angle_offset), np.sin(angle_offset)]) * cluster_offset
        for i, node in enumerate(nodes):
            angle = 2 * np.pi * i / len(nodes)
            pos[node] = cluster_center + np.array([np.cos(angle), np.sin(angle)]) * 0.5  # Adjust node spacing within cluster

    return pos

def remove_indices(matrix, row_indices, col_indices):
    reduced_matrix = [row for index, row in enumerate(matrix) if index not in row_indices]
    reduced_matrix = [[item for index, item in enumerate(row) if index not in col_indices] for row in reduced_matrix]
    
    return np.array(reduced_matrix)

def remove_by_indices(original_list, indices_to_remove):
    indices_set = set(indices_to_remove)
    return [element for index, element in enumerate(original_list) if index not in indices_set]

non_ms = [2, 6, 7, 11, 16, 17, 20, 24, 25, 26, 30, 36, 37, 40] #memcache, redis, mongodb
ms_slist = remove_by_indices(s_list, non_ms)

service_type_mapping = {i: extract_service_type(label) for i, label in enumerate(ms_slist)}
service_clusters = defaultdict(list)
for node, service_type in service_type_mapping.items():
    service_clusters[service_type].append(node)

color_cycle = itertools.cycle(plt.cm.tab20.colors)  # Use a colormap with enough colors
service_type_colors = {service_type: next(color_cycle) for service_type in service_clusters.keys()}

fig, axes = plt.subplots(1, 2, figsize=(15, 10), dpi=150)
for ax, (title, matrix) in zip(axes, network_centroids.items()):
    np.fill_diagonal(matrix, 0)
    matrix = remove_indices(matrix, non_ms, non_ms)
    matrix = topcorr.tmfg(np.abs(matrix))
    matrix = nx.to_numpy_array(matrix)
    G = nx.from_numpy_array(matrix)
    pos = initial_positions(service_clusters)

    fixed_pos = pos.copy()  # Keep initial positions for fixed nodes
    pos = nx.spring_layout(G, pos=pos, fixed=fixed_pos.keys(), weight='weight', iterations=50)

    for service_type, nodes in service_clusters.items():
        node_positions = np.array([pos[node] for node in nodes])
        cluster_center = node_positions.mean(axis=0)
        radius = np.linalg.norm(node_positions - cluster_center, axis=1).max() + 0.2
        circle = Circle(cluster_center, radius, color=service_type_colors[service_type], alpha=0.2, zorder=0)
        ax.add_patch(circle)

    node_colors = []
    for node in G.nodes():
        service_type = service_type_mapping[node]
        node_colors.append(service_type_colors[service_type])
    weights = np.array([d['weight'] for u, v, d in G.edges(data=True)])
    scaled_weights = 4 * weights / weights.max()  # Scaling factor for better visualization
    node_sizes = 500 
    node_edge_color = 'k'

    nx.draw(G, pos, ax=ax, with_labels=False, node_size=node_sizes, 
            node_color=node_colors, edge_color='k', width=scaled_weights, 
            linewidths=2, edgecolors=node_edge_color)

    ax.set_title(title)

legend_elements = [Circle((0, 0), 0.5, color=color, label=service_type) for service_type, color in service_type_colors.items()]
fig.legend(handles=legend_elements, loc='lower center', ncol=7, fontsize=15, frameon=False)

plt.tight_layout(rect=[0, 0.1, 1, 0.95])
plt.subplots_adjust(bottom=0.2)
plt.show()
```

```python
def plot_time_series_and_clusters(data, indices, cluster_labels):
    fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
    
    ax[0].plot(indices, data[:, :]*100, linestyle='-')
    ax[0].set_title('CPU Usage')
    ax[0].set_ylabel('CPU Usage (%)')

    n_clusters = np.unique(cluster_labels).size
    cmap = get_cmap('viridis', n_clusters)
    colors = {label: cmap(i) for i, label in enumerate(np.unique(cluster_labels))}

    prev_label = cluster_labels[0]
    start_index = indices[0]
    for i in range(1, len(cluster_labels)):
        if cluster_labels[i] != prev_label or i == len(cluster_labels) - 1:
            end_index = indices[i] if cluster_labels[i] != prev_label else indices[i] + 1
            ax[1].axvspan(start_index, end_index, color=colors[prev_label], alpha=0.5)
            start_index = indices[i]
            prev_label = cluster_labels[i]
    
    ax[1].set_title('Cluster Assignments Over Time')
    ax[1].set_xlabel('Index')
    ax[1].set_xlim(indices[0], indices[-1])

    plt.tight_layout()
    plt.show()
    
plot_time_series_and_clusters(X[361:,:],range(len(labels)), labels)
```
